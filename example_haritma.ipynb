```python
path1='/Users/legolason/study/mesfit/'                  # the path of the source code file and qsopar.fits
path2='/Users/legolason/study/mesfit/test/data/haritma/' # path of fitting results
path3='/Users/legolason/study/mesfit/test/data/haritma/'   # path of figure 
path4='/Users/legolason/study/sed/dustmap/'             # path of dusp reddening map

#Requried
# an important note that all the data input must be finite, especically for the error !!!
data=fits.open(path1+'test/data/haritma/spectrum.fits') 

lam=data[1].data['wavelength'][15:-15]        # OBS wavelength [A]
flux=data[1].data['flux'][15:-15]            # OBS flux [erg/s/cm^2/A]
err=data[1].data['err'][15:-15]  # 1 sigma error
z=1.42               # Redshift
'''
plt.figure(figsize=(12,5))
plt.plot(lam[15:-15]/2.42,flux[15:-15])
plt.plot([5007,5007],[-1,1],'k--')
plt.plot([4861,4861],[-1,1],'k--')
plt.plot([6563,6563],[-1,1],'k--')
plt.ylim(-0.4,0.5)
'''

# get data prepared 
q = QSOFit(lam, flux, np.ones_like(lam)/10., z, path = path1)

start = timeit.default_timer()
# do the fitting
q.Fit(name = 'spectrum',nsmooth = 1, and_or_mask = False, deredden = False, reject_badpix = False, wave_range = None,\
      wave_mask =None, decomposition_host = False, BC03=False, Mi = None, npca_gal = 5, npca_qso = 20, \
      Fe_uv_op = False, poly = False, BC = False, rej_abs = False, initial_guess = None, MC = False, \
      n_trails = 5, linefit = True, tie_lambda = True, tie_width = True, tie_flux_1 = True, tie_flux_2 = True,\
      save_result = False, plot_fig = True,save_fig = True, plot_line_name = True, plot_legend = True, \
      dustmap_path = path4, save_fig_path = path3, save_fits_path = path2,save_fits_name = None)

end = timeit.default_timer()
print 'Fitting finished in : '+str(np.round(end-start))+'s'
# grey shade on the top is the continuum windiows used to fit.

print q.line_result_name[12:18]
fwhm,sigma,ew,peak,area = q.line_prop(5007,q.line_result[12:18],'broad') 
print("OIII line:")
print("FWHM (km/s)", fwhm)
print("Sigma (km/s)", sigma)
print("EW (A)",ew)
print("Peak (A)",peak)
print("area (10^(-17) erg/s/cm^2)",area)

print q.line_result_name[18:21]
fwhm,sigma,ew,peak,area = q.line_prop(6563,q.line_result[18:21],'broad') 
print("Ha line:")
print("FWHM (km/s)", fwhm)
print("Sigma (km/s)", sigma)
print("EW (A)",ew)
print("Peak (A)",peak)
print("area (10^(-17) erg/s/cm^2)",area)
```
